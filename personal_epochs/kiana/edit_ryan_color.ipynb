```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import lightkurve as lk
#import os.path
#import csv

df = pd.read_csv(r'C:\Users\Kiana\OneDrive\Documents\GitHub\sunnyhills\data\current\full_key.csv')
df=df.dropna()
df

rightAscension = df['GDR2_RA'][:500]
declination = df['GDR2_DEC'][:500]
distances = np.array(df['Distance'][:500])

stellarColors = df['bp-rp'][:500]
stellarMags = df['absolute_mag'][:500]
b=stellarMags-(5*np.log10(distances)+5)
a=stellarColors
#flip x and y axis
ax = plt.scatter(a,b,c=np.log10((distances[:500])),cmap='plasma',zorder=2)
plt.xlabel("Stellar Color")
plt.ylabel("Stellar Magnitude")
plt.gca().invert_yaxis()

c = plt.colorbar()
c.set_label("Log Distance From Earth")

l = plt.legend(loc='upper right')
l.set_zorder(20)  # put the legend on top
plt.show()
```
