```python
import lightkurve as lk
import pyarrow.feather as feather
import pandas as pd
import matplotlib.pyplot as plt
import astropy
from astropy.table import TableMergeError
from astropy.utils.metadata import MergeConflictError
```

```python
%matplotlib inline
```

```python
def download_lightcurve(tic_id):
    tic_id = 'TIC ' + tic_id
    result = lk.search_lightcurve(tic_id,mission='TESS',cadence='long')
    lightcurves = result.download_all(quality_bitmask='hardest')
    try:
        lightcurves.plot(normalize=True)
        lc = lightcurves.stitch()
        lc.plot()
        feather.write_feather(lc.to_pandas(),'C:/Users/60002/Documents/GitHub/sunnyhills/data/current/stitched_curves/' + tic_id, compression='lz4')
    except (AttributeError, MergeConflictError, TableMergeError):
        print('no light curve found')
    
```

```python
def download_best(tic_id):
    tic_id = 'TIC ' + tic_id
    result = lk.search_lightcurve(tic_id,mission='TESS',cadence='long')
    try:
        lightcurves = result.download(quality_bitmask='hardest')
        lightcurves.plot()
        feather.write_feather(lightcurves.to_pandas(),'C:/Users/60002/Documents/GitHub/sunnyhills/data/current/best_curves/' + tic_id, compression='lz4')
    except AttributeError:
        pass
    
```
