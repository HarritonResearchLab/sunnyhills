```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
from tqdm import tqdm 

counts = []
diffs = []

lc_dir = r'C:\Users\Research\Documents\GitHub\sunnyhills\data\current\processed\two_min_lightcurves'
for file in tqdm(os.listdir(lc_dir)):
    if file!='.gitkeep': 
        df = pd.read_csv(lc_dir+'\\'+file)
        counts.append(len(df['cleaned_time']))
        diffs.append(len(df['trend_time'])-len(df['trend_flux']))
```

```python
plt.hist(counts)
plt.gca().set(xscale='log', yscale='log', xlabel='log(observations)', ylabel='log(frequency)')
```

```python
plt.hist(counts, bins=20)
plt.gca().set(yscale='log', xlabel='observations', ylabel='log(frequency)')
```

```python
counts = np.array(counts)
mask = np.array(counts)<10**5

fig, axs = plt.subplots(1,2, figsize=(6,3))

plt.tight_layout()

ax = axs[0]
ax.hist(counts[mask])

ax = axs[1]
ax.hist(counts[~mask])

for i in [0,1]: 
    axs[i].set(xlabel='Usable Observations', ylabel='Frequency')
```
