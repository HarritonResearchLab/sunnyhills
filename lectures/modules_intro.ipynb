```python
import pandas as pd # module for dataframes; nickname is pd
import matplotlib.pyplot as plt # module for making plots; nickname is plt
```

## Simple Scatter Plot

```python
x_list = [1, 2, 3, 4, 5]
y_list = [2, 4, 6, 8, 10]

plt.plot(x_list,y_list, lw=10) # lw changes line width; I don't generally change lw or point size
plt.show()
```

## Scatter Plot (no line between points)

```python
plt.scatter(x_list, y_list)
plt.show()
```

## Simple Customization

```python
plt.scatter(x_list, y_list, color='cornflowerblue', label='random numbers', s=400) # change color; give line a label; change point size

plt.legend() # make legend

# label the plot

plt.xlabel('X') # xlabel
plt.ylabel('Y') # ylabel
plt.title('First plot') # title

plt.show()
```

## Histogram

```python
z_list = [12, 1, 2, 3, 4, 5, 2, 2, 2, 1, 2, 3, 4, 5, 6, 4, 3, 2, 5, 7, 8, 8, 4, 2]

plt.hist(z_list, color='mediumseagreen')
plt.show()

# make x label X
# make y label Frequency 
# make title Histogram 
# change color 
```
