```python
import pandas as pd
import numpy as np
```

```python
### PANDAS MERGE
df = pd.read_csv(r'C:\Users\Research\Documents\GitHub\sunnyhills\lectures\test.txt')

first_arr_names = ['name','orbital_period','radius']

second_arr_names = ['name', 'ra','dec','star_distance']

names, orbital_periods,radius = (np.array(df[arr_name]) for arr_name in first_arr_names)

names, ra, dec, star_distance = (np.array(df[arr_name]) for arr_name in second_arr_names)

zipped_1 = list(zip(names, orbital_periods,radius))
df1 = pd.DataFrame(zipped_1, columns=first_arr_names)
df1.to_csv('df1.csv', index=False)

zipped_2 = list(zip(names, ra, dec, star_distance))

df2 = pd.DataFrame(zipped_2, columns=second_arr_names)
df2 = df.sample(frac=1)
df2.to_csv('df2.csv', index=False)
```
