### Savitzkyâ€“Golay Filter

```python
import numpy as np
import pandas as pd
import scipy
from scipy.signal import savgol_filter 
import matplotlib.pyplot as plt 
from scipy.interpolate import interp1d

plt.rcParams['font.family']='serif' 
plt.style.use('ggplot') 

initial_df = pd.read_csv('https://raw.githubusercontent.com/HarritonResearchLab/sunnyhills/main/lectures/references/lotka-voltara.csv')
```

```python
hare_years = np.array(initial_df['Year'])
lynx_years = hare_years

hare_counts = np.array(initial_df['Hare'])
lynx_counts = np.array(initial_df['Lynx'])

fig, ax = plt.subplots()

ax.plot(hare_years, lynx_counts)
ax.scatter(hare_years, lynx_counts, label='Lynx')
ax.plot(hare_years, hare_counts)
ax.scatter(hare_years, hare_counts, label='Hare')

ax.set(xlabel='Year', ylabel='Count')

ax.legend()

plt.show()
```

### Interpolation

```python
x = np.arange(-4*np.pi, 4*np.pi, 1.5)
y = 4*np.sin(x)+np.random.normal(loc=0, scale=0.01, size=x.shape)

y += 8

fig, ax = plt.subplots()

ax.scatter(x,y)
ax.plot(x,y)

ax.set(xlabel='x', ylabel='y')

plt.show()
```

```python
f = interp1d(x, y, kind='cubic', fill_value='extrapolate')

x_new = np.arange(-3.9*np.pi,3.9*np.pi,0.1)
y_new = f(x_new)


fig, ax = plt.subplots()

ax.scatter(x,y)
ax.plot(x_new, y_new)

ax.set(xlabel='x', ylabel='y')

plt.show()
```

### Flux Normalization

$y'=\frac{y}{\mathrm{mean}(y)}$

```python
x = np.arange(-4*np.pi, 4*np.pi, 1.5)
y = 4*np.sin(x)+np.random.normal(loc=0, scale=0.01, size=x.shape)

y += 8

y_prime = y/np.mean(y)

fig, ax = plt.subplots()

ax.scatter(x, y_prime)
ax.plot(x, y_prime)

ax.set(xlabel='x', ylabel='y (normalized)')

plt.show()
```

### Signals in Noise

```python
x1 = np.arange(0, 4, 0.25)
y1 = np.array([1]*len(x1))

y2 = np.array([])

for x_index, x_val in enumerate(x1): 
    val = y1[x_index]
    if 1 <= x_val <=3: 
        y2 = np.append(y2,(x_val-2)**2) 

    else: 
        y2 = np.append(y2, val)

fig, ax = plt.subplots()

ax.scatter(x1, y2)

ax.set(xlabel='x', ylabel='y')

plt.show()
```
